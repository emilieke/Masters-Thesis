# Data Visualization

Import packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)
color = sns.color_palette()
```

Load data

```python
dtrain = pd.read_csv('Data/dtrain_cleaned.csv')
dtest = pd.read_csv('Data/dtest_cleaned.csv')
```

```python
dtrain = dtrain.drop('Unnamed: 0',1)
dtest = dtest.drop('Unnamed: 0',1)
dtrain.describe()
```

#### Age distribution

```python
with sns.plotting_context('notebook',font_scale=1.3):
    plt.figure(figsize=(16,16))
    sns.set_style('whitegrid')
    
    plt.subplot(2,1,1)
    sns.countplot(x='Alder',data=dtrain,alpha=0.9)
    sns.plt.title('Age Distribution Train Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('Age',fontsize=16)

    plt.subplot(2,1,2)
    sns.countplot(x='Alder',data=dtest,alpha=0.9)
    sns.plt.title('Age Distribution Test Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('Age',fontsize=16)
    plt.savefig('age_dist')
    plt.show()
```

```python
age_group_train=[]
for age in dtrain['Alder']:
    if(age<=26):
        age_group_train.append('18-25')
    if(26<age<=34):
        age_group_train.append('26-34')
    if(34<age<=41):
        age_group_train.append('35-41')
    if(41<age<=48):
        age_group_train.append('42-48')
    if(48<age<=55):
        age_group_train.append('49-55')
    if(55<age<=62):
        age_group_train.append('56-62')
    if(62<age<=70):
        age_group_train.append('63-70')
```

```python
age_group_test=[]
for age in dtest['Alder']:
    if(age<=26):
        age_group_test.append('18-25')
    if(26<age<=34):
        age_group_test.append('26-34')
    if(34<age<=41):
        age_group_test.append('35-41')
    if(41<age<=48):
        age_group_test.append('42-48')
    if(48<age<=55):
        age_group_test.append('49-55')
    if(55<age<=62):
        age_group_test.append('56-62')
    if(62<age<=70):
        age_group_test.append('63-70')
```

```python
with sns.plotting_context("notebook",font_scale=1.5):
    plt.figure(figsize=(16,6))
    sns.set_style('whitegrid')
    
    plt.subplot(1,2,1)
    sns.countplot(x=age_group_train,alpha=0.9,
                 order=['18-25','26-34','35-41','42-48','49-55','56-62','63-70'])
    sns.plt.title('Age Group Distribution Train Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('Age groups',fontsize=16)
    
    plt.subplot(1,2,2)
    sns.countplot(x=age_group_test,alpha=0.9,
                 order=['18-25','26-34','35-41','42-48','49-55','56-62','63-70'])
    sns.plt.title('Age Group Distribution Test Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('Age groups',fontsize=16)
    
    plt.savefig('age_group_dist')
    plt.show()
```

#### Target variables distribution
The 16 product variables present in this dataset are the following:

1.  AntAktAvtBank: Banking
2.  AntAktAvtInnskudd: Deposits
3.  AntAktAvtKredittkort: Credit
4.  AntKredittkortDebet: Debit card
5.  AntKredittkortKreditt: Credit card
6.  AntSparekonto: Savings account
7.  AntBrukskonto: Current account
8.  AntAvtBSU: Youth house savings account
9.  AntAktiveSkadefors: Accident insurance
10. AntAktiv_AndreForsikringer: Other insurances
11. AntAktSpareforsikring: Savings insurance
12. AntAktUtlanAvt: Loan
13. AntAktBoliglan: Mortgage
14. DNBAntAktBoliglan: DNB Mortgage
15. AntAktGaranterte: Guarantees
16. AntAktLink: Links

```python
target_cols = ['AntAktAvtBank',
 'AntAktAvtInnskudd',
 'AntAktAvtKredittkort',
 'AntKredittkortDebet',
 'AntKredittkortKreditt',
 'AntSparekonto',
 'AntBrukskonto',
 'AntAvtBSU',
 'AntAktiveSkadefors',
 'AntAktiv_AndreForsikringer',
 'AntAktSpareforsikring',
 'AntAktUtlanAvt',
 'AntAktBoliglan',
 'DNBAntAktBoliglan',
 'AntAktGaranterte',
 'AntAktLink']

target_dtrain = dtrain[target_cols]
target_dtest = dtest[target_cols]
    
target_dtrain.columns = ['Banking','Deposits','Credit','Debit_card','Credit_card','Savings_account','Current_account',
                         'Youth_house_savings','Accident_insurance','Other_insurances','Savings_insurance','Loan',
                         'Mortgage','DNB_mortgage','Guarantees','Links']
target_dtest.columns = ['Banking','Deposits','Credit','Debit_card','Credit_card','Savings_account','Current_account',
                        'Youth_house_savings','Accident_insurance','Other_insurances','Savings_insurance','Loan',
                        'Mortgage','DNB_mortgage','Guarantees','Links']
```

```python
target_counts_dtrain = target_dtrain.astype('float64').sum(axis=0)
target_counts_dtest = target_dtest.astype('float64').sum(axis=0)

with sns.plotting_context("notebook",font_scale=1.3):
    plt.figure(figsize=(16,6))
    sns.set_style('whitegrid')
    
    plt.subplot(1,2,1)
    sns.barplot(target_counts_dtrain.index,target_counts_dtrain.values,alpha=0.9)
    sns.plt.title('Distribution of Products Train Set',fontsize=18)
    plt.xlabel('Product Name',fontsize=16)
    plt.ylabel('Count',fontsize=16)
    plt.xticks(rotation='vertical')
    
    plt.subplot(1,2,2)
    sns.barplot(target_counts_dtest.index,target_counts_dtest.values,alpha=0.9)
    sns.plt.title('Distribution of Products Test Set',fontsize=18)
    plt.xlabel('Product Name',fontsize=16)
    plt.ylabel('Count',fontsize=16)
    plt.xticks(rotation='vertical')
    
    plt.savefig('product_dist')
    plt.show()
```

```python
dtrain['FylkeNr'] = dtrain['FylkeNr'].astype('int64')
dtest['FylkeNr'] = dtest['FylkeNr'].astype('int64')

county_train=[]
for f in range(len(dtrain['FylkeNr'])):
    if(dtrain['FylkeNr'][f]>13):
        val = dtrain['FylkeNr'][f]-1
    else:
        val = dtrain['FylkeNr'][f]
    county_train.append(val)

county_test=[]
for f in range(len(dtest['FylkeNr'])):
    if(dtest['FylkeNr'][f]>13):
        val = dtest['FylkeNr'][f]-1
    else:
        val = dtest['FylkeNr'][f]
    county_test.append(val)
```

```python
with sns.plotting_context('notebook',font_scale=1.5):
    plt.figure(figsize=(16,6))
    sns.set_style('whitegrid')

    plt.subplot(1,2,1)
    sns.countplot(x=county_train)
    sns.plt.title('County Distribution Train Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('County groups',fontsize=16)
    
    plt.subplot(1,2,2)
    sns.countplot(x=county_test)
    sns.plt.title('County Distribution Test Set',fontsize=18)
    plt.ylabel('Count',fontsize=16)
    plt.xlabel('County groups',fontsize=16)
    
    plt.savefig('county_dist')
    plt.show()
```

#### Number of product versus customer lifetime

```python
with sns.plotting_context('notebook',font_scale=1.5):
    plt.figure(figsize=(16,6))
    sns.set_style('whitegrid')

    plt.subplot(1,2,1)
    sns.boxplot(x='LevetidKundeforholdDg', 
                y='AntallProdukter', 
                data=dtrain, 
                orient='h',
                order=[9,8,7,6,5,4,3,2,1,0])
    sns.plt.title('Customer Lifetime for Number of Products Train Set',fontsize=18)
    plt.ylabel('Number of Products',fontsize=16)
    plt.xlabel('Customer Lifetime (Days)',fontsize=16)
    
    plt.subplot(1,2,2)
    sns.boxplot(x='LevetidKundeforholdDg', 
                y='AntallProdukter', 
                data=dtest, 
                orient='h',
                order=[9,8,7,6,5,4,3,2,1,0])
    sns.plt.title('Customer Lifetime for Number of Products Test Set',fontsize=18)
    plt.ylabel('Number of Products',fontsize=16)
    plt.xlabel('Customer Lifetime (Days)',fontsize=16)
    
    plt.savefig('lifetime_product_boxplot')
    plt.show()
```

#### Total number of products by age

```python
data_products_train = dtrain.groupby(['Alder'])['AntallProdukter'].mean()
data_products_test = dtest.groupby(['Alder'])['AntallProdukter'].mean()

with sns.plotting_context('notebook',font_scale=1.3):
    plt.figure(figsize=(16,16))
    sns.set_style('whitegrid')

    plt.subplot(2,1,1)
    sns.barplot(data_products_train.index,data_products_train.values,alpha=0.9)
    sns.plt.title('Mean Number of Products by Age Train Set',fontsize=18)
    plt.ylabel('Total number of Products',fontsize=16)
    plt.xlabel('Age',fontsize=16)
    
    plt.subplot(2,1,2)
    sns.barplot(data_products_test.index,data_products_test.values,alpha=0.9)
    sns.plt.title('Mean Number of Products by Age Test Set',fontsize=18)
    plt.ylabel('Total number of Products',fontsize=16)
    plt.xlabel('Age',fontsize=16)
    plt.savefig('mean_product_age')
    plt.show()
```
